# **Group by and Aggregating**

```python
import pandas as pd
```

```python
df = pd.read_csv(r'/content/Flavors.csv')
df
```

```python
group_by_dataframe = df.groupby('Base Flavor')
```

```python
group_by_dataframe.mean()
```

```python
df.groupby('Base Flavor').mean()
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').sum()
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean','max','count','sum'], 'Texture Rating':['mean','max','count','sum'] })
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating': ['mean','max','count','sum']})
```

```python
df.groupby('Base Flavor').describe()
```

# **Merge, Join, and Concatenate**

```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/LOTR.csv')
df1
```

```python
df2 = pd.read_csv('/content/LOTR 2.csv')
df2
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.join(df2, how = 'outer',on = 'FellowshipID', lsuffix = '_left', rsuffix = '_right')
```

```python
df3 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left',rsuffix = '_Right', how = 'outer')
df3
```

```python
df1.append(df2)
```

```python
pd.concat([df1, df2], join = 'outer', axis = 1)
```
